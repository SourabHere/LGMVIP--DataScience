# Importing libraries

```python
import cv2
import matplotlib.pyplot as plt 
```

## Getting image

```python
img=cv2.imread("dog.jpg")
```

```python
plt.imshow(img)
plt.axis(False)
plt.show()
```

## Converting BGR to RGB 

```python
new_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(new_img)
plt.axis(False)
plt.show()
```

## Converting BGR to Gray image 

```python
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img)
plt.axis(False)
plt.show()
```

## Inverting image colors 

```python
invert_img=cv2.bitwise_not(gray_img)
plt.imshow(invert_img)
plt.axis(False)
plt.show()
```

## Applying Gausian blur in image

```python
blur_img=cv2.GaussianBlur(invert_img, (111,111),0)
plt.imshow(blur_img)
plt.axis(False)
plt.show()
```

# Again inverting blurred image

```python
invblur_img=cv2.bitwise_not(blur_img)
plt.imshow(invblur_img)
plt.axis(False)
plt.show()
```

## Obtaining sketch image 

```python
sketch_img=cv2.divide(gray_img,invblur_img, scale=256.0)
plt.imshow(sketch_img)
plt.axis(False)
plt.show()
```

# Finally plotting original vs sketched image

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Original image', size=18)
plt.imshow(new_img)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Sketch', size=18)
rgb_sketch=cv2.cvtColor(sketch_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
