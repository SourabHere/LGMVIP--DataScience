# Importing required libraries 

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier  
from sklearn import tree
```

# Loading dataset

```python
iris=pd.read_csv("Iris.csv")
iris.head()
```

```python
iris.Species.value_counts()
```

```python
labels=iris["Species"].copy()

features=iris.drop("Species",axis=1)
```

```python
labels
```

```python
features=features.drop("Id",axis=1)
features
```

```python
iris.info()
```

```python
iris.describe()
```

```python
len(features),len(labels)
```

# Dividing train and test dataset into 80% train and 20% test

```python
from sklearn.model_selection import train_test_split

train_feature_set,test_feature_set,train_label_set,test_label_set=train_test_split(features,labels,test_size=0.2,random_state=42)
```

```python
len(train_label_set),len(test_label_set),len(train_feature_set),len(test_feature_set)
```

## Setting model 

```python
model=DecisionTreeClassifier()
```

## Training our model 

```python
model.fit(train_feature_set,train_label_set)
```

## Predicting using trained model

```python
predictions=model.predict(test_feature_set)
```

```python
for i in range(0,len(predictions)):
    print("Actual:",np.array(test_label_set)[i],"   ","Predicted:",predictions[i])
```

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
print(classification_report(test_label_set,predictions))
```

```python
accuracy_score(test_label_set,predictions)
```

## Visualizing Decision Trees

```python
text_representation = tree.export_text(model)
print(text_representation)
```

```python
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (10,10), dpi=300)
tree.plot_tree(model);
```

```python
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa', 'versicolor', 'virginica']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (10,10), dpi=300)
tree.plot_tree(model,
               feature_names = fn, 
               class_names=cn,
               filled = True,);
fig.savefig('imagename.png')
```
